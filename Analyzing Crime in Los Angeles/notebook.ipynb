### Explore the crimes.csv dataset and use your findings to answer the following questions:

```python
# Re-run this cell
#Â Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
crimes = pd.read_csv("crimes.csv", parse_dates=["Date Rptd", "DATE OCC"], dtype={"TIME OCC": str})
crimes.head()
```

### Question 1: Identifying Peak Crime Hour

**Objective:** Determine the hour with the highest frequency of crimes and store the result as an integer variable called `peak_crime_hour`.


```python
# Start coding here
# Use as many cells as you need
crimes['hour occ'] = np.floor(crimes['TIME OCC'].astype(int) / 100)
```

```python
peak_crime_hour = int(crimes['hour occ'].value_counts().index[0])
peak_crime_hour
```

### Question 2: Identifying Area with the Largest Frequency of Night Crimes

**Objective:** Determine the area with the largest frequency of night crimes (committed between 10pm and 3:59am) and save the result as a string variable called `peak_night_crime_location`.

**Solution:**
We converted the 'TIME OCC' column to an integer format and created a new column, 'TIME OCC INT'. Then, we filtered crimes that occurred between 10pm and 3:59am and identified the area with the largest frequency of night crimes.

```python
crimes['TIME OCC INT'] =crimes['TIME OCC'].astype(int)
```

```python
peak_night_crime_location = crimes[(crimes['TIME OCC INT'] >= 2200) | (crimes['TIME OCC INT'] <=359)]['AREA NAME'].value_counts().index[0]
peak_night_crime_location
```

### Question 3: Analyzing Crimes by Victim Age Groups

**Objective:** Identify the number of crimes committed against victims by age group (0-17, 18-25, 26-34, 35-44, 45-54, 55-64, 65+). Save the results as a pandas Series called `victim_ages`.

**Solution:**
We defined age bins and labels, then created a new column, 'age_group', using `pd.cut()` to bin victim ages into discrete intervals. The resulting Series, `victim_ages`, displays the frequency of crimes for each age group.

```python
age_bins = [0, 17, 25, 34, 44, 54, 64, np.inf]
age_labels = ["0-17", "18-25", "26-34", "35-44", "45-54", "55-64", "65+"]

# Add a new column using pd.cut() to bin values into discrete intervals
crimes["age_group"] = pd.cut(crimes["Vict Age"],
                               bins=age_bins,
                               labels=age_labels)

# Find the category with the largest frequency
victim_ages = crimes["age_group"].value_counts()
victim_ages
```
